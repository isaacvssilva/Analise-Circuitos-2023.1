```python
# importando lib
import pandas as pd
import matplotlib.pyplot as plt

# importando dataset que contem os dados coletados durante o experimento
data = pd.read_csv('./dados_sensor_peso.csv')
data = pd.DataFrame(data)
```

```python
# pegando as colunas importantes para fazer a plotagem dos graficos
# e analisar os dados
tensao_entrada_3_3v = data[data['Tensao_entrada (V)'] == 3.3]

tensao_entrada_5v = data[data['Tensao_entrada (V)'] == 5]

tensao_entrada_9v = data[data['Tensao_entrada (V)'] == 9]

variacao_trimpot_3_3v = data['Resistencia_Trimpot (Ω)'][0:49]
variacao_trimpot_5v = data['Resistencia_Trimpot (Ω)'][50:99]
variacao_trimpot_9v = data['Resistencia_Trimpot (Ω)'][100:149]

ddp_3_3v = data['DDP (mV)'][0:49]
ddp_5v = data['DDP (mV)'][50:99]
ddp_9v = data['DDP (mV)'][100:149]
```

```python
# Plotando o gráfico
plt.figure(figsize=(15, 10))  # Define o tamanho da figura
plt.plot(variacao_trimpot_3_3v, ddp_3_3v, linestyle='solid', marker='o', linewidth=2, color='blue', label='Dados 1')

# Configurando o título e os rótulos dos eixos
plt.title('Relação entre Resistência e DDP - Tensão de 3.3V', fontsize=14)
plt.xlabel('Variação Trimpot (Ω)', fontsize=12)
plt.ylabel('DDP (mV)', fontsize=12)
plt.grid(True)  # Adiciona uma grade de fundo
plt.legend()  # Adiciona uma legenda

# Exibindo o gráfico
plt.show()
```

```python
# Plotando o gráfico
plt.figure(figsize=(15, 10))  # Define o tamanho da figura
plt.plot(variacao_trimpot_5v, ddp_5v, linestyle='solid', marker='o', linewidth=2, color='blue', label='Dados 1')

# Configurando o título e os rótulos dos eixos
plt.title('Relação entre Resistência e DDP - Tensão de 5V', fontsize=14)
plt.xlabel('Variação Trimpot (Ω)', fontsize=12)
plt.ylabel('DDP (mV)', fontsize=12)
plt.grid(True)  # Adiciona uma grade de fundo
plt.legend()  # Adiciona uma legenda

# Exibindo o gráfico
plt.show()
```

```python
# Plotando o gráfico
plt.figure(figsize=(15, 10))  # Define o tamanho da figura
plt.plot(variacao_trimpot_9v, ddp_9v, linestyle='solid', marker='o', linewidth=2, color='blue', label='Dados 1')

# Configurando o título e os rótulos dos eixos
plt.title('Relação entre Resistência e DDP - Tensão de 9V', fontsize=14)
plt.xlabel('Variação Trimpot (Ω)', fontsize=12)
plt.ylabel('DDP (mV)', fontsize=12)
plt.grid(True)  # Adiciona uma grade de fundo
plt.legend()  # Adiciona uma legenda

# Exibindo o gráfico
plt.show()
```
